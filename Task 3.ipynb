```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("D:/Intership Python/Task 2/01.Data Cleaning and Preprocessing.csv")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Bin Y-Kappa values
df['Y-Kappa Bin'] = pd.cut(df['Y-Kappa'], bins=10)

# Aggregate Blow Flow by Y-Kappa Bin
agg_df = df.groupby('Y-Kappa Bin')['BlowFlow'].mean().reset_index()

# Bar Chart: Aggregated Y-Kappa vs Blow Flow with edges
plt.figure(figsize=(12, 6))
plt.bar(agg_df['Y-Kappa Bin'].astype(str), agg_df['BlowFlow'], color='blue', edgecolor='black')
plt.xlabel('Y-Kappa')
plt.ylabel('Blow Flow')
plt.title('Bar Chart: Aggregated Y-Kappa vs Blow Flow')
plt.xticks(rotation=45)
plt.show()

# Scatter Plot: Y-Kappa vs Blow Flow
plt.figure(figsize=(12, 6))
plt.scatter(df['Y-Kappa'], df['BlowFlow'], color='blue')
plt.xlabel('Y-Kappa')
plt.ylabel('Blow Flow')
plt.title('Scatter Plot: Y-Kappa vs Blow Flow')
plt.show()

# Line Chart: Y-Kappa vs Blow Flow
plt.figure(figsize=(12, 6))
plt.plot(df['Y-Kappa'], df['BlowFlow'], marker='o', linestyle='-', color='red')
plt.xlabel('Y-Kappa')
plt.ylabel('Blow Flow')
plt.title('Line Chart: Y-Kappa vs Blow Flow')
plt.show()

# Hypothetical data for Sales and Products
sales_data = {
    'Products': ['Product A', 'Product B', 'Product C', 'Product D'],
    'Sales': [150, 200, 300, 250]
}
sales_df = pd.DataFrame(sales_data)

# Bar Chart: Sales vs Products with edges
plt.figure(figsize=(12, 6))
plt.bar(sales_df['Products'], sales_df['Sales'], color='green', edgecolor='black')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Bar Chart: Sales vs Products')
plt.show()

# Line Chart: Sales vs Products
plt.figure(figsize=(12, 6))
plt.plot(sales_df['Products'], sales_df['Sales'], marker='o', linestyle='-', color='purple')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Line Chart: Sales vs Products')
plt.show()
```
